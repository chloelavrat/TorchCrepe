```python
import torch
import torch.nn as nn
from crepe.utils import activation_to_frequency, frequency_to_activation
```

```python
sr = 16000
test_freq = torch.arange(0., float(sr//2), 1).unsqueeze(0)
```

```python
# 
activation = frequency_to_activation(test_freq[0,:])
freq_out = activation_to_frequency(activation)
```

```python
# plot 
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1)  # Create two subplots, one above the other

%matplotlib inline
plt.figure(figsize=(10, 4))  # Adjust the figure size

# First plot on the first set of axes
ax1.plot(test_freq[0, :], color='b')
ax1.set_ylabel('Test Frequency')
ax1.set_xlabel('X-Axis for Test Frequency')

# Second plot on the second set of axes
ax2.plot(freq_out, color='r')
ax2.set_ylabel('Frequency Out')
ax2.set_xlabel('X-Axis for Frequency Out')

plt.show()
```
